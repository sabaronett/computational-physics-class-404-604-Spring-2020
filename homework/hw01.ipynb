### Equation to Plot

\begin{equation}
f(x)=\frac{\pi}{2}+2 \left[ sin(x)+\frac{sin(3x)}{3}+\frac{sin(5x)}{5}+....+\frac{sin(nx)}{n} \right]
\tag{1}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, n):
    """
    Function to plot (see Eq. 1).
    
    Parameters
    ----------
    x : float or array_like
        Independent variable(s)
    n : int
        Number of sine terms in series (see Eq. 1)
    
    Returns
    -------
    Value of f(x) given some n (Eq. 1).    
    """
    if not isinstance(n, int) or n<1:
        return 0
    else:
        series = 0
        for i in range(n):
            series += np.sin((i+1)*x)/(i+1)
    return np.pi/2 + 2*series
```

### First with $n$ = 1, 2, 3

```python
x = np.arange(-np.pi, np.pi, 0.01)
y1 = f(x, 1)
y2 = f(x, 2)
y3 = f(x, 3)

xticks = np.arange(-np.pi, np.pi+np.pi/4, np.pi/4)
xlabels = [r'$-\pi$',r'$-3\pi/4$',r'$-\pi/2$',
           r'$-\pi/4$',0,r'$\pi/4$',
           r'$\pi/2$',r'$3\pi/4$',r'$\pi$']
yticks = np.arange(-2*np.pi, 2*np.pi+np.pi/4, np.pi/4)
ylabels = [r'$-2\pi$',r'$-7\pi/8$',r'$-3\pi/2$',r'$-5\pi/4$',
           r'$-\pi$',r'$-3\pi/4$',r'$-\pi/2$',r'$-\pi/4$',0,
           r'$\pi/4$',r'$\pi/2$',r'$3\pi/4$',r'$\pi$',
           r'$5\pi/4$',r'$3\pi/2$',r'$7\pi/8$',r'$2\pi$',]

plt.figure(figsize=[5,10])
plt.title(r'$f(x)$ for various $n$')
plt.xlabel(r'$x$')
plt.xlim(-np.pi,np.pi)
plt.xticks(xticks, xlabels)
plt.ylabel(r'$f(x)$')
plt.ylim(-2*np.pi,2*np.pi)
plt.yticks(yticks, ylabels)
plt.grid()
plt.plot(x, y1, label=r'$n=1$')
plt.plot(x, y2, label=r'$n=2$')
plt.plot(x, y3, label=r'$n=3$')
plt.legend(loc="best")
plt.show()
```

### Now with $n$ = 11, 21, 31

```python
y11 = f(x, 11)
y21 = f(x, 21)
y31 = f(x, 31)

plt.figure(figsize=[5,10])
plt.title(r'$f(x)$ for various $n$')
plt.xlabel(r'$x$')
plt.xlim(-np.pi,np.pi)
plt.xticks(xticks, xlabels)
plt.ylabel(r'$f(x)$')
plt.ylim(-2*np.pi,2*np.pi)
plt.yticks(yticks, ylabels)
plt.grid()
plt.plot(x, y11, label=r'$n=11$')
plt.plot(x, y21, label=r'$n=21$')
plt.plot(x, y31, label=r'$n=31$')
plt.legend(loc="best")
plt.show()
```
